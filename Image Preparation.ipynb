```python
import matplotlib.pyplot as plt
import numpy as np
import time
import skimage
import skimage.io
import utils.image_utils as iu
```

```python
#Import an image
#Note: This requires the dataset to be located in the current folder

image = skimage.io.imread('./semantic_drone_dataset/original_images/000.jpg')
window_size = 1000
```

```python
#Try out some rotation stuff (and time it)

a = time.time()
rotated = skimage.transform.rotate(image[:,:4000], 90,resize=True)
b = time.time()
rotated_np = np.rot90(image[:,:4000])
c = time.time()
fig, ax = plt.subplots(2)
ax[0].imshow(rotated)
ax[1].imshow(rotated_np)
print(b-a)
print(c-b)
```

```python
#Create the patches and plot some of them

d = time.time()
all_patches = iu.prepare_image(image, 1000, True)
e = time.time()
print(len(all_patches))
print(e-d)

fig, ax = plt.subplots(2,4)
ax[0,0].imshow(all_patches[0])
ax[0,1].imshow(all_patches[1])
ax[0,2].imshow(all_patches[2])
ax[0,3].imshow(all_patches[3])
ax[1,0].imshow(all_patches[124])
ax[1,1].imshow(all_patches[125])
ax[1,2].imshow(all_patches[126])
ax[1,3].imshow(all_patches[127])
```
